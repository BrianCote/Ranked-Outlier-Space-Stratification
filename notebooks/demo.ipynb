```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets

os.chdir('..')
from ROSS import ross
```

## Simple test on 2D data, for easy visualization.

```python
plt.rcParams["figure.figsize"] = (10,10)
plt.style.use('seaborn-dark-palette')

arr = datasets.make_blobs(n_samples=1000, n_features=2, centers=3, cluster_std=1.2, random_state=800)[0]
noise = np.array([np.random.uniform(low=min(arr[:,0]), high=max(arr[:,0]), size=25),
                  np.random.uniform(low=min(arr[:,1]), high=max(arr[:,1]), size=25)]).T
arr = np.vstack([arr, noise])
plt.scatter(arr[:,0], arr[:,1])
```

```python
df = pd.DataFrame(arr)
df['labels'] = ross.fit_predict(arr)

plt.scatter(df[0], df[1], c=df['labels'], cmap='turbo')
```
